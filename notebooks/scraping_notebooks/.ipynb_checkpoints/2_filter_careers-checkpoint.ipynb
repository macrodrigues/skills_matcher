## Import the CSV (pre-cleaned)

```python
import numpy as np
import pandas as pd
```

```python
jobs = pd.read_csv('jobs_scraped.csv')
```

```python
jobs = jobs.set_index(['major_ISCO', 'major_jobs']).drop(['Unnamed: 0'], axis = 1).reset_index()
```

## Filter Data Frame by careers

```python
from ast import literal_eval
jobs['careers_in_major'] = jobs['careers_in_major'].apply(literal_eval)
jobs = jobs.explode('careers_in_major')
```

```python
jobs = jobs.drop(['sub_major_ISCO','unit_group_ISCO'], axis = 1)
jobs = jobs.rename({'major_ISCO': 'ISCO'}, axis = 1)
jobs = jobs[['careers_in_major', 'ISCO', 'major_jobs', 'openings']].reset_index().drop(['index'], axis = 1)
```

```python
jobs = jobs.drop(labels = list(jobs.loc[jobs.ISCO == 43].index), axis = 0) # drop 43 careers
jobs = jobs.drop(labels = list(jobs.loc[jobs.ISCO == 81].index), axis = 0) # drop 81 careers
jobs = jobs.drop(labels = list(jobs.loc[jobs.ISCO == 94].index), axis = 0) # drop 94 careers
jobs = jobs.drop(labels = list(jobs.loc[jobs.ISCO == 44].index), axis = 0) # drop 44 careers
jobs = jobs.drop(labels = list(jobs.loc[jobs.ISCO == 41].index), axis = 0) # drop 41 careers
jobs = jobs.drop(labels = list(jobs.loc[jobs.ISCO == 82].index), axis = 0) # drop 82 careers
jobs = jobs.drop(labels = list(jobs.loc[jobs.ISCO == 96].index), axis = 0) # drop 96 careers
jobs = jobs.drop(labels = list(jobs.loc[jobs.ISCO == 73].index), axis = 0) # drop 73 careers
```
