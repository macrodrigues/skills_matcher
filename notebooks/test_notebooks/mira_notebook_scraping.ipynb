```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
data = pd.read_csv('../data/ISCO_TO_SOC.csv')
data = list(data.SOC)
data = [number.strip() for number in data]
data = [f'{number}.00' for number in data]
```

```python
code = '17-2031.00'
```

```python
def fetch_page(code):
    """Takes an ISCO code as input and returns the HTML soup
    by scraping the Tucareers website"""
    onet_url = f"https://www.onetonline.org/link/summary/{code}#Skills"
    response = requests.get(onet_url)
    soup = BeautifulSoup(response.content, "html.parser", from_encoding="utf-8")
    return soup
```

```python
def get_skills_tech(code):
    
    soup = fetch_page(code)
    
    skills = []
    for element in soup.find_all('div', class_= "section_TechnologySkills"):
        if element.find_all('li'):
            for skill in element.find_all('li'):
                skills.append(skill.get_text())

    skills = [skills[index].replace('\n\n', '') for index, skill in enumerate(skills)]
    skills = [re.sub("[\(\[].*?[\)\]]", "", skills[index]) for index, skill in enumerate(skills)]
    skills = [skills[index].replace('Minitab', '') for index, skill in enumerate(skills)]
    skills = [skills[index].split(';') for index, skill in enumerate(skills)]
    skills = [item for sublist in skills for item in sublist] # from list in list to list
    skills = [skills[index].strip() for index, skill in enumerate(skills)]
    skills = [skills[index].split('—') for index, skill in enumerate(skills)]
    skills = [item for sublist in skills for item in sublist] # from list in list to list
    skills = list(filter(None, skills)) # remove empty strings
    skills = [skills[index].strip() for index, skill in enumerate(skills)]
    
    return skills
```

```python
tech_skills = []

for number in data:
    tech_skills.append(get_skills_tech(number))
```

```python
tech_skills_final = [item for sublist in tech_skills for item in sublist]
```

```python
df_tech_skills_final = pd.DataFrame(tech_skills_final)
```

```python
df_tech_skills_final.to_csv('../tech_skills.csv')
```

```python
def get_skills_general(code):
    
    soup = fetch_page(code)
    
    skills = []
    for element in soup.find_all('div', class_= "section_Skills"):
        if element.find_all('li'):
            for skill in element.find_all('b'):
                skills.append(skill.get_text())

    skills = [skills[index].split('-') for index, skill in enumerate(skills)]
    
    return skills
```

```python
all_skills_general = []

i = 0

for number in data:
    all_skills_general.append(get_skills_general(number))
    print(i)
    i += 1
```

```python
len(all_skills_general)
```

```python
all_skills_general_final = [item for sublist in all_skills_general for item in sublist]
```

```python
df_all_skills_general_final = pd.DataFrame(all_skills_general_final)
df_all_skills_general_final.to_csv('../general_skills.csv')
```

```python
def get_knowledge(code):
    
    soup = fetch_page(code)
    
    skills = []
    for element in soup.find_all('div', class_= "section_Knowledge"):
        if element.find_all('li'):
            for skill in element.find_all('b'):
                skills.append(skill.get_text())

    skills = [skills[index].split('-') for index, skill in enumerate(skills)]
    
    return skills
```
