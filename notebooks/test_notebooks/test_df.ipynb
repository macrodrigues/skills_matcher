```python
import pandas as pd
```

```python
df_auto = pd.read_csv('data_only_auto.csv')
df_manual = pd.read_csv('data_only_manual.csv')
```

```python
df_auto = df_auto.rename(columns = {'Unnamed: 0': 'index'})
df_manual = df_manual.rename(columns = {'Unnamed: 0': 'index'})
```

```python
df_manual
```

```python
data_final = df_auto.merge(df_manual[['index', 'entities_manual_label']], on = 'index', how = 'inner')
```

```python
data_final
```

```python
for index, row in data_final.iterrows():
    if row.entities_auto_label == '()' and row.entities_manual_label == '[]':
        data_final.drop(index, inplace=True)
```

```python
data_final = data_final.reset_index().drop(['level_0', 'index'], axis = 1)
```

```python
data_final.to_csv('data_final_raw.csv')
```
